```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x =  np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO approach for matplotlib
# create figure object and call methods of the object
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.2,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# now putting two set of figure in the same plot

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_xlabel('x1_label')
axes1.set_ylabel('y1_label')
axes1.set_title('large plot')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_xlabel('x1_label')
axes2.set_ylabel('y1_label')
axes2.set_title('large plot')

```

```python
# Figure Size and DPI
fig,axes = plt.subplots(figsize=(8,2))
#ax = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)


axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi= 200)
```

```python
# Now let us discuss about the legend
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,x**2)
axes.plot(x,x**3)
```

```python
# now we want to create a legend for both the lines
# for making a legend we first have to put the label in both the curve
# So, inside the plot function we have to add a new argument called the label
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X-Square')
ax.plot(x,x**3,label='X-Cube')
ax.legend(loc=0)
```

```python
#Plot Appearance
# linewith 
# line style
# alpha: Transperancy of the line
# 

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='black',linewidth=1,alpha = 0.5,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',
         markeredgewidth=3,markeredgecolor='green')
```
