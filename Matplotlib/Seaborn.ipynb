```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# let us plot the distribution plot of a univariate data
# the solid line is known as kde which is kernel density estimation, we can remove it by saying kde = False
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind= 'hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind= 'reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind= 'kde')
```

```python
# Pairplot gives the relation between all the variables in the dataset
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# Code to build kernel density estimation plot
sns.kdeplot(tips['total_bill'])
```
