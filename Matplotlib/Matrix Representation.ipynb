```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
# if we need to present the actual value, we can use the annotation method

sns.heatmap(tc,annot=True)
```

```python
# What if we need to change the colour of the heatmap
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
# how to convert a non-pivotal view of a pivotal view
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp)
```

```python
# add more arguments
sns.heatmap(fp, cmap='coolwarm',linecolor='white',linewidths=2)
```

```python
# note that we can use standard_scale to normalise the scale from 0 to 1

sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
